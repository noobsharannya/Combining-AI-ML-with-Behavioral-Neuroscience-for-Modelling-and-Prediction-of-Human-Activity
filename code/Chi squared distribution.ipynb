```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi2
```

```python
x = np.linspace(0, 20, 400)
degrees_of_freedom = [2,4,6,8,10]
for k in degrees_of_freedom:
    plt.plot(x, chi2.pdf(x, k), label=f'k={k}')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.title('Chi-Squared Distribution for Different Degrees of Freedom (k)')
plt.legend()
plt.show()
```
