```python
test_scores={}
test1=['EI','WI','EC','WC','EA','WA']
test2=['AM','LC','MC','ND','PfW','SE']
test3=['E','A','C','N','O']
```

```python
l=df.shape[0]
t1_scores=[]
t2_scores=[]
t3_scores=[]
bxi_scores=[]
cxi_scores=[]
txi_scores=[]
final=[]
for i in range(l):
    t1=0
    x=df['EI'][i]
    if(x>=6 and x<=7):
        t1+=10
    elif(x>=2 and x<=5):
        t1+=8
    else:
        t1+=6
    x=df['WI'][i]
    if(x>=6 and x<=7):
        t1+=10
    elif(x>=2 and x<=5):
        t1+=8
    else:
        t1+=6
    x=df['EC'][i]
    if(x>=4 and x<=7):
        t1+=10
    elif(x>=2 and x<=3):
        t1+=8
    else:
        t1+=6
    x=df['WC'][i]
    if(x>=4 and x<=7):
        t1+=10
    elif(x>=2 and x<=3):
        t1+=8
    else:
        t1+=6
    x=df['EA'][i]
    if(x>=4 and x<=6):
        t1+=10
    elif(x>=7 and x<=9):
        t1+=8
    else:
        t1+=6
    x=df['WA'][i]
    if(x>=4 and x<=6):
        t1+=10
    elif(x>=7 and x<=9):
        t1+=8
    else:
        t1+=6
    t1_scores.append(t1)
    t2=0
    for y in test2:
        x=df[y][i]
        if(x>=4 and x<=5):
            t2+=10
        elif(x>=2.6 and x<=3.9):
            t2+=8
        else:
            t2+=6
    t2_scores.append(t2)
    t3=0
    x=df['E'][i]
    if(x<=25):
        t3+=10
    elif(x<=40):
        t3+=8
    else:
        t3+=6
    x=df['A'][i]
    if(x<=30):
        t3+=10
    elif(x<=40):
        t3+=8
    else:
        t3+=6
    x=df['C'][i]
    if(x<=25):
        t3+=6
    elif(x<=35):
        t3+=8
    else:
        t3+=10
    x=df['N'][i]
    if(x<=25):
        t3+=10
    elif(x<=40):
        t3+=8
    else:
        t3+=6
    x=df['O'][i]
    if(x<=30):
        t3+=10
    elif(x<=40):
        t3+=8
    else:
        t3+=6
    t3_scores.append(t3)
    bxi=2*t1+t2
    bxi_scores.append(bxi)
    bxi=(bxi-108)/(180-108)
    cxi=2*t1+t3
    cxi_scores.append(cxi)
    cxi=(cxi-102)/(170-102)
    txi=2*t1+0.5*df['AM'][i]+0.5*df['PfW'][i]+df['O'][i]
    txi_scores.append(txi)
    txi=(txi-84)/(140-84)
    f=((bxi+cxi+txi)/3)*(100)
    final.append(f)
```

```python
test_scores['test1']=t1_scores
test_scores['test2']=t2_scores
test_scores['test3']=t3_scores
test_scores['BXI']=bxi_scores
test_scores['CXI']=cxi_scores
test_scores['TXI']=txi_scores
test_scores['composite']=final
```
