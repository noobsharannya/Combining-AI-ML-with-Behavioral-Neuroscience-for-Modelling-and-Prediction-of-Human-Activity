```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import re
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neural_network import MLPClassifier
from sklearn.svm import SVC
from sklearn.cluster import KMeans
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, silhouette_score,normalized_mutual_info_score,adjusted_rand_score
```

```python
test1_traits=['EI','WI','EC','WC','EA','WA']
test2_traits=['AM','LC','ND','MC','PfW','SE']
test3_traits=['E','A','C','N','O']
```

```python
features=pd.read_csv('CSV/features.csv')
com=pd.read_csv('CSV/data.csv')
data=pd.read_csv('CSV/raw_data.csv')
```

```python
def test(start,num_features):
    codelist=[]
    for i in range(start,start+num_features):
        codelist.append(features['Code'][i])
    return codelist
```

```python
test1=test(0,9)
test1=test1+test(9,9)
test1=test1+test(18,9)
test1=test1+test(27,9)
test1=test1+test(36,9)
test1=test1+test(45,9)
test1_rows=[i for i in data.index if data['Code'][i] in test1]
```

```python
test2=test(54,3)
test2=test2+test(57,4)
test2=test2+test(61,4)
test2=test2+test(65,2)
test2=test2+test(67,3)
test2=test2+test(70,4)
test2_rows=[i for i in data.index if data['Code'][i] in test2]
```

```python
test3=test(74,10)
test3=test3+test(84,10)
test3=test3+test(94,10)
test3=test3+test(104,10)
test3=test3+test(114,10)
test3_rows=[i for i in data.index if data['Code'][i] in test3]
```

```python
mapping={
     'most people':3,
    'many people':2,
    'some people':1,
    'a few people':0,
    'one or two people':-1,
    'nobody':-2,
    'usually': 3,
    'often':2,
    'sometimes':1,
    'occasionally':0,
    'rarely':-1,
    'never':-2,
    'disagree':-2,
    'slightly disagree':-1,
    'neutral':0,
    'slightly agree':1,
    'agree':2
}
```

```python
mean=[]
for i in data.index:
    s=0
    c=0
    for column in data.columns[2:]:
        val=data[column][i]
        res = re.sub(' +', ' ',val)
        res=res.lower()
        res=''.join(c for c in res if (c.isalpha()or c==' '))
        res=res.strip()
        if(i>=54 and i<=73 and not pd.isna(val)):
            s=s+int(val)
            c=c+1
        elif(res in mapping):
            s=s+mapping[res]
            c=c+1
    if(c==0):
        print(data.loc[i])
    s=s/c
    mean.append(s)
```

```python
def compute_features(df,feature_mapping,rows):
    X=[]
    for column in df.columns[2:]:
        values=list(df[column])
        l=len(values)
        x=[]
        for i in rows:
            value=values[i]
            res = re.sub(' +', ' ',value)
            res=res.lower()
            res=''.join(y for y in res if (y.isalpha()or y==' '))
            res=res.strip()
            if(i>=54 and i<=73 and not pd.isna(value)): 
                x.append(int(value))
            elif(res in feature_mapping):
                x.append(feature_mapping[res])
            else:
                x.append(mean[i])
        X.append(x)
    return np.array(X)
```

```python
X1=compute_features(data,mapping,test1_rows)
X2=compute_features(data,mapping,test2_rows)
X3=compute_features(data,mapping,test3_rows)
```

```python
def label12(x):
    if(x>=54):
        return 0
    elif(x>42):
        return 1
    else:
        return 2
def label3(x):
    if(x>=46):
        return 0
    elif(x>=36):
        return 1
    else:
        return 0
```

```python
Y1=[label12(x) for x in list(com['Test-1'])]
Y2=[label12(x) for x in list(com['Test-2'])]
Y3=[label3(x) for x in list(com['Test-3'])]
```

```python
k_values=[]
scores=[]
for k in range(2,10):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X1)
    labels = kmeans.labels_
    k_values.append(k)
    scores.append(silhouette_score(X1, labels))
plt.plot(k_values,scores)
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X1, Y1, test_size=0.1, random_state=42)
svm_classifier = SVC(kernel='rbf', C=1.0, random_state=42)
svm_classifier.fit(X_train, y_train)
y_pred = svm_classifier.predict(X_train)
print("Support Vector Machines: ")
print("Training accuracy: "+str(accuracy_score(y_train, y_pred)))
y_pred = svm_classifier.predict(X_test)
print("Testing accuracy: "+str(accuracy_score(y_test, y_pred)))
random_forest_classifier = RandomForestClassifier(n_estimators=100, random_state=42)
random_forest_classifier.fit(X_train, y_train)
y_pred = random_forest_classifier.predict(X_train)
print("Random Forest Classifier :")
print("Training accuracy: "+str(accuracy_score(y_train, y_pred)))
y_pred = random_forest_classifier.predict(X_test)
print("Testing accuracy: "+str(accuracy_score(y_test, y_pred)))
logistic_regression_classifier = LogisticRegression(max_iter=1000, random_state=42)
logistic_regression_classifier.fit(X_train, y_train)
y_pred = logistic_regression_classifier.predict(X_train)
print("Logistic Regression Classifier :")
print("Training accuracy: "+str(accuracy_score(y_train, y_pred)))
y_pred = logistic_regression_classifier.predict(X_test)
print("Testing accuracy: "+str(accuracy_score(y_test, y_pred)))
mlp = MLPClassifier(hidden_layer_sizes=(100,50), max_iter=500, random_state=42)
mlp.fit(X_train, y_train)
y_pred = mlp.predict(X_train)
print("Multi Layer Perceptron :")
print("Training accuracy: "+str(accuracy_score(y_train, y_pred)))
y_pred = mlp.predict(X_test)
print("Testing accuracy: "+str(accuracy_score(y_test, y_pred)))
```

```python
def distribution(clusters,labels,num):
    distribution=np.zeros((num,num))
    for (c,l) in zip(clusters,labels):
        distribution[c][l]=distribution[c][l]+1
    return distribution
```
